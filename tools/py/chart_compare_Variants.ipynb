```python
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects
from matplotlib.lines import Line2D

# Example data
categories = ['Variante 1', 'Variante 2', 'Variante 3', 'Variante 4']
values = [11.3, 25.5, 10.3, 3.7]

bar_width = 0.4  

# Create figure and axis
fig, ax = plt.subplots(figsize=(6, 4))

# Create a simple, elegant bar chart
ax.bar(categories, values, width=bar_width, color='gray', edgecolor='black', linewidth=0.8, alpha=0.9)

# Add labels and title
#ax.set_xlabel("Categories", fontsize=12)
ax.set_ylabel("Anzahl erstellter Belege\nin einem Werktag (Millionen)", fontsize=12)
#ax.set_title("Anzahl der erstellten Belege in einem Arbeitstag", fontsize=14, fontweight="bold")

# Add a horizontal line at y = 1 in dark red
ax.axhline(y=1, color='darkred', linestyle='--', linewidth=1.5)

ax.set_ylim(0, 29)

# Custom legend: Add a dark red line with explanation
legend_line = Line2D([0], [0], color='darkred', lw=2, linestyle='--')
ax.legend([legend_line], ['Anforderung'], loc='upper left')

# Beautify the layout
plt.tight_layout()

plt.savefig(r'D:\ba\Codes\arbeit\pics\vars.png')

# Show plot
plt.show()
```
